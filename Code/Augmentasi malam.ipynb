```python
for i in range(275):
  im = Image.open("/content/drive/MyDrive/skenny/data/"+str(i)+".jpg")

  enhancer = ImageEnhance.Color(im)
  im = enhancer.enhance(0)
  enhancer = ImageEnhance.Contrast(im)
  im = enhancer.enhance(0.85)
  enhancer = ImageEnhance.Brightness(im)
  im = enhancer.enhance(0.35)
  enhancer = ImageEnhance.Sharpness(im)
  im = enhancer.enhance(0.0)

  im.save("/content/drive/MyDrive/skenny/data/"+str(i+275)+".jpg")

  #!cp /content/drive/MyDrive/skenny/labelme/{str(i)}.json /content/drive/MyDrive/skenny/datanight/{str(i+275)}.json
  im.close()
```

```python
import json

for i in range(275,550):
  with open(f'/content/drive/MyDrive/skenny/data/{i}.json', 'r+') as f:
      data = json.load(f)
      data["imagePath"] = f"{i}.jpg" # <--- add `id` value.
      f.seek(0)        # <--- should reset file position to the beginning.
      json.dump(data, f, indent=2)
      f.truncate()     # remove remaining part
      print(i)
!clear
```
